```python
import pandas as pd
from sklearn import linear_model 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
year = df[['year']]
year.head()
```

```python
income = df[['per capita income (US$)']]
income.head()
```

```python
plt.scatter(year,income)
plt.xlabel('year')
plt.ylabel('per capita income')
```

```python
model = linear_model.LinearRegression()
model.fit(year,income)
```

```python
pred = model.predict([[2020]])
pred = pred.tolist()
pred[0]
```

```python
plt.plot(df.year,model.predict(df[['year']]))
```

```python
year.loc[year.size] = 2020
income.loc[income.size] = pred[0]
```

```python
plt.scatter(year,income)
plt.plot(year,model.predict(year))
```
